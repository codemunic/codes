```python
def data_processing_train(df):
    columns=['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area','Loan_Status']
    df=df[columns]
    df['Gender']=np.where(df['Gender']=='Male',1,0)
    df['Married']=np.where(df['Married']=='Yes',1,0)
    df['Education']=np.where(df['Education']=='Graduate',1,0)
    df['Self_Employed']=np.where(df['Self_Employed']=='Yes',1,0)
    df['Loan_Status']=np.where(df['Loan_Status']=='Y',1,0)
    df['Property_Area'].replace({'Urban':1,'Rural':2,'Semiurban':3},inplace=True)
    df=df.replace({np.NaN:0})
    df['Dependents'].replace({0:0,'0':0,'1':1,'2':2,'3+':3},inplace=True)
    columns_X=['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']
    X=df[columns_X]
    y=df['Loan_Status']
    return X,y
```
