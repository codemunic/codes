```python
df['Age']=df['Age'].dt.days
df['Date between Booking and Flight']=df['Date between Booking and Flight'].dt.days
```

```python
columns=['Age','Date between Booking and Flight','Booking date(ID)','Flight Date(ID)','Flight Time(ID)','Distance','Class[ID]']
```

```python
Data=df[columns]
```

```python
Data.head(10)
```

```python
Data.shape
```

```python
y=df['Fare']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(Data,y,random_state=0)
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
linreg=LinearRegression().fit(X_train_scaled,y_train)
```

```python
linreg.score(X_train_scaled,y_train)
```

```python
linreg.score(X_test_scaled,y_test)
```

```python
linridge=Ridge(alpha=20).fit(X_train_scaled,y_train)
```

```python
linridge.score(X_train_scaled,y_train)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
mlpreg=MLPRegressor(hidden_layer_sizes=[100,50,25],activation='relu',alpha=0.1,solver='lbfgs').fit(X_train_scaled,y_train)
```

```python
mlpreg.score(X_train_scaled,y_train)
```

```python
mlpreg.score(X_test_scaled,y_test)
```
